```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
from matplotlib import rcParams
plt.style.use('ggplot')
rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
df = pd.read_csv('BUND_matched_edits.csv', index_col=0)
df['year_month'] = pd.to_datetime(df['year_month'], format='%Y-%m-%d')
```

```python
grouped = df[['year_month', 'total_act', 'total_acts_surv_48h', 'num_edits']].groupby('year_month').sum()
```

```python
grouped.head()
```

```python
plt.plot(grouped['num_edits'], label='Officials')
plt.title('Number of edits made by Officials')
plt.xlabel('Time')
plt.ylabel('Number of edits(Revisions)')
plt.legend()
plt.show()
```
